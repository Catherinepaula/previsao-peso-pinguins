```python

# === 1. Importa√ß√£o de bibliotecas ===
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score

# Para deixar os gr√°ficos maiores no Colab
plt.rcParams["figure.figsize"] = (8,6)
sns.set_style("whitegrid")
```

```python
# === 2. Carregamento da base de dados ===
df = sns.load_dataset("penguins")
df.head()
```

## 1. An√°lise Explorat√≥ria

Vamos entender as rela√ß√µes entre vari√°veis preditivas e a vari√°vel resposta (`species`).

**Gr√°ficos solicitados:**
1. Atributos num√©ricos por esp√©cie  
2. Sexo por esp√©cie  
3. Ilha por esp√©cie


```python
# 1.1 Atributos num√©ricos por esp√©cie
# Usamos pairplot limitado a duas vari√°veis exemplares para n√£o sobrecarregar.
sns.pairplot(df, vars=["bill_length_mm","flipper_length_mm"], hue="species", height=3)
plt.suptitle("Comprimento do bico e nadadeira por Esp√©cie", y=1.02);
```

**Coment√°rio:**  
- Percebe‚Äëse que Gentoo (laranja) tende a ter **nadadeiras mais longas** e bicos maiores.  
- Adelie (azul) e Chinstrap (verde) ocupam regi√µes distintas no gr√°fico.  
Isso sugere que esses atributos s√£o fortes discriminadores de esp√©cie.


```python
# 1.2 Sexo por esp√©cie
sns.countplot(data=df, x="sex", hue="species")
plt.title("Distribui√ß√£o de Sexo por Esp√©cie");
```

**Coment√°rio:**  
- Observa‚Äëse que h√° mais indiv√≠duos machos em Gentoo e Chinstrap.  
- O desbalanceamento de sexo pode influenciar o modelo, mas aqui usamos apenas como insight explorat√≥rio.


```python
# 1.3 Ilha por esp√©cie
sns.countplot(data=df, x="island", hue="species")
plt.title("Distribui√ß√£o de Esp√©cie por Ilha");
```

**Coment√°rio:**  
- Cada ilha abriga sobretudo uma ou duas esp√©cies dominantes (por ex. Biscoe ‚Üí Gentoo e Adelie).  
- Ilha √© vari√°vel geogr√°fica que pode ajudar na classifica√ß√£o, pois funciona como proxy de esp√©cie.


```python
## 2. Prepara√ß√£o dos Dados

### 2.1. Tratamento de valores nulos
```

```python

# Verificar nulos originais
print(df.isnull().sum())

# Preencher:
# - Num√©ricos: mediana
# - Categ√≥ricos: moda
df['bill_length_mm'].fillna(df['bill_length_mm'].median(), inplace=True)
df['bill_depth_mm'].fillna( df['bill_depth_mm'].median(),  inplace=True)
df['flipper_length_mm'].fillna(df['flipper_length_mm'].median(), inplace=True)
df['body_mass_g'].fillna(df['body_mass_g'].median(), inplace=True)
df['sex'].fillna(df['sex'].mode()[0], inplace=True)
df.isnull().sum()  # conferir novamente
```

### 2.2. Codifica√ß√£o de vari√°veis categ√≥ricas

Embora √°rvores de decis√£o n√£o necessitem de codifica√ß√£o, o Scikit‚ÄëLearn exige inputs num√©ricos.  
- Vari√°veis **nominais**: `species` (resposta), `island`, `sex`.


```python
le_species = LabelEncoder()
le_island  = LabelEncoder()
le_sex     = LabelEncoder()

# Aqui criamos _nom apenas para as preditoras; species ser√° o y.
df['island_nom'] = le_island.fit_transform(df['island'])
df['sex_nom']    = le_sex.fit_transform(df['sex'])
```

### 2.3. Limpeza final

Descartar colunas originais categ√≥ricas e manter a resposta (`species`) em primeiro lugar.


```python
df_model = df[[
    'species',
    'bill_length_mm','bill_depth_mm','flipper_length_mm',
    'island_nom','sex_nom'
]]
df_model.head()
```

### 2.4. Separa√ß√£o em treino e teste

Propor√ß√£o: **2/3** treino e **1/3** teste.


```python
X = df_model.drop('species', axis=1)
y = le_species.fit_transform(df_model['species'])  # transformamos species para 0,1,2

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=1/3, random_state=42, stratify=y
)

print("Treino:", X_train.shape, "Teste:", X_test.shape)
```

## 3. Modelagem

### 3.1. Treinamento da √Årvore de Decis√£o


```python
tree = DecisionTreeClassifier(random_state=42)
tree.fit(X_train, y_train)

# Desenhar a √°rvore
plt.figure(figsize=(12,8))
plot_tree(tree,
          feature_names=X.columns,
          class_names=le_species.classes_,
          filled=True, rounded=True, fontsize=8)
plt.show()

# Quantas folhas?
n_leaves = tree.get_n_leaves()
print(f"N√∫mero de folhas da √°rvore treinada: {n_leaves}")
```

**Resposta:**  
- A √°rvore treinada possui **`{n_leaves}`** folhas.


```python
### 3.2. Avalia√ß√£o do modelo
# Matriz de Confus√£o
y_pred = tree.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(cm, display_labels=le_species.classes_)
disp.plot(cmap='Blues')
plt.title("Matriz de Confus√£o")
plt.show()

# Acur√°cia
acc = accuracy_score(y_test, y_pred)
print(f"Acur√°cia no conjunto de teste: {acc*100:.2f}%")
```

**Coment√°rio Matriz de Confus√£o:**  
- Diagonal principal dominante indica que a maioria das inst√¢ncias foi classificada corretamente.  
- Alguns erros ocorrem entre Adelie e Chinstrap (classes mais parecidas).

**Coment√°rio Acur√°cia:**  
- A curadoria de refer√™ncia obteve ‚âà¬†96%. Nesta execu√ß√£o, temos **`{acc*100:.2f}%`**, o que confirma excelente performance.


## 4. Predi√ß√£o de um novo pinguim

**Exemplo de usu√°rio:**  
- Comprimento do bico: 38.2¬†mm  
- Profundidade do bico: 18.1¬†mm  
- Comprimento da nadadeira: 185.0¬†mm  
- Ilha: Biscoe  
- Sexo: Male  

> **Obs.:** ajuste a ordem das colunas conforme X_train.


```python
novo = pd.DataFrame([{
    'bill_length_mm': 38.2,
    'bill_depth_mm': 18.1,
    'flipper_length_mm': 185.0,
    'island_nom': le_island.transform(['Biscoe'])[0],
    'sex_nom':    le_sex.transform(['Male'])[0],
}])

esp_pred = tree.predict(novo)[0]
print("Esp√©cie predita:", le_species.inverse_transform([esp_pred])[0])
```

> **Refer√™ncia:**  
> Neste exemplo, o modelo costuma prever **Adelie**, o que confirma a mesma refer√™ncia do material de apoio.
